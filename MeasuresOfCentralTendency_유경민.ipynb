# Measures of central tendency (중심경향도)
<br>> 중심경향도은 전체적인 평균이나 데이터 셋의 평균 값을 설명하기 위한 것

*   mean(샘플값의 평균)
*   median(중간)
*   mode(가장 많은 빈도로 나타난 숫자)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/hands-on-exploratory-data-analysis-with-python/master/Chapter%205/data.csv")
df.head()
```

```python
height = df["height"]
mean = height.mean()
median = height.median()
mode = height.mode()
print(mean, median, mode)
```
