## 8.1 차원의 저주 / 8.2 차원 축소를 위한 접근방법

8.1 차원의 저주

- 차원이 높아질수록 데이터는 sparse하게 분포해지면서 데이터가 부족 => 학습이 어려워지는 현상
  * 모든 점들의 학습을 위해 모든 구역을 확인해야 함
  * 같은 구역내 점들은 구별 불가(1-Dimension의 경우..)

* 매니폴드
   - 고차원 공간안의 샘플들이 저차원 다양체(manifold)형태로 분포해 있다고 가정하고 차원을 줄이는 기법
   - 고차원상에서 가까운 거리에 있던 데이터 포인트들일지라도, 매니폴드를 보다 저차원 공간으로 맵핑하면 오히려 거리가 멀어질 수 있는 특징
   - 저차원의 공간상에서 가까운 점끼리는 실제로도 비슷한 특징feature을 가짐
