# 7.6 스태킹
1. 개념
 - 모든 예측기의 예측을 취합하는 간단한 함수(직접 투표) 같은 기존 방식대신 **예측을 취합하는 모델을 훈련시키는 방식**
 - 취합하는 모델 자체를 훈련 시키는 모델

image.png

[방식]
1. 각 예측 모델에서 다음 결과 (3.1, 2.7, 2.9) 예측
2. 마지막 예측기(블렌딩)에서 예측을 입력받아 최종 (3.0)을 예측

- 블렌더를 학습시키는 일반적인 방법은 hold-out 세트를 사용

image.png

[방식]
1. 먼저 training set을 2개의 subset으로 나눔
2. subset1은 첫번째 레이어의 예측을 훈련 시키는 데 사용
3. 훈련된 첫번째 레이어의 예측기로 subset2에 대한 예측을 생성
  (subset 이지만 훈련에 사용이 안됐으므로 test 셋으로 사용 가능)

  image.png

4. 홀드 아웃 세트의 각 샘플에 세 개의 예측값 존재
5. 타깃값은 그대로 쓰고 앞에서 예측한 3개의 값을 입력 변수로 사용하는 새로운 훈련 세트 생성
6. 블렌더가 새로운 훈련 세트로 학습
 (즉, 첫번째 레이어의 예측 3개를 이용해 타깃값을 예측하도록 학습)

* 각 모델별로 (하나는 선형회귀, 다른 하나는 랜덤 포레스트 회귀)로 훈련 시키는 것도 가능.
1) subset을 3개로 나누어, 
2) subset1은 첫번째 레이어의 예측기 훈련
3) subset2로 첫번째 예측기들의 예측을 만들어 블렌더 레이어의 예측기를 위한 훈련 세트를 생성
4) subset3으로 두번째 예측기들의 예측을 만들어 세번째 레이어를 훈련시키기 위한 훈련 세트를 만드는 데 사용
