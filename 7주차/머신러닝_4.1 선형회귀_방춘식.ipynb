# 4.1.1 정규 방정식

비용함수를 최소화하는 theta값을 찾기 위한 해석적 방벙이 있음.

이를 정규 방정식이라고 함

theta_best = (XtX)-1 Xt y

```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100,1)
```

정규방정식을 사용해 theta_best를 계산하겠으며, 넘파이 선형대수 모듈(np.linalg)에 있는 inv()함수를 사용해 역행렬을 계산하고 dot() 메소드를 사용해 행렬 곱셉을 계산

```python
X_b = np.c_[np.ones((100, 1)), X]  # 모든 샘플에 x0= 1을 추가합니다.
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.figure(figsize=(20, 10))
plt.rc('font', size=20)
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

사이킷런에서 선형회귀를 수행하는 것은 간담함
