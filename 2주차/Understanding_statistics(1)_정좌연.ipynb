# 연속확률 분포 vs 이산확률분포

## 1. 이산확률분포 : 이산변수들의 확률분포, 확률질량함수(probability mass function, PMF)로 표현
- 1) "떨어져 있다~"라는 뜻의 이산.  ex) 주사위 숫자 확률 변수 X에서 X= 1,2,3,4,5,6 등
- 2) 이산확률변수의 확률변수가 가질 수 있는 값을 셀 수 있는 경우
- PMF는 한 개의 동전을 두 번 던지는 시행에서 앞면이 나오는 횟수를 확률변수 X라고 할 때, 확률변수 X가 취할 수 있는 값은 0, 1, 2이며,
- 이때 각각의 확률은 \begin{matrix} P(X=0) = \frac{1}{4} \\ P(X=1) = \frac{2}{4} = \frac{1}{2} \\ P(X=2) = \frac{1}{4} \end{matrix}

- 유형: 이산균등분포, 푸아송분포, 베르누이 분포, 기하분포, 초기하분포, 이항분포, 음의 이항 분포, 다향 분포
    
## 2. 연속확률분포 : 확률밀도함수(PDF)를 이용해 분포를 표현할 수 있는 경우
- 1) 확률변수가 연속해서 이어져 있는 경우로, 셀 수 없는 확률변수를 가진 확률분포
- 예) 고등학생의 키 확률변수 X,  160cm~170cm 사이에는 무수히 많은 변수가 있음 (160.1,160.12...)
- 정규분포, 연속균등분포, 카이제곱분포, 감마분포


## 확률밀도함수(PDF, probability density function)
- 연속확률분포를 표현한 함수
- ex) 150센치~160센치의 학생들의 비율 즉, 확률을 구하고 싶다면,
확률밀도함수 f(x)와 X축 그리고 X=150 ~ X=160 사이의 넓이 
- P(150 ≤ X ≤ 160) = 0.4

- 확률밀도함수에서 변수의 시작을 a / 변수의 끝을 b 라고 하면, P(a ≤ X ≤ b) = 무조건 1





# 3. Uniform distribution (균등분포, 연속균등분포)
- 분포가 특정 범위 내에서 균등하게 나타나 있을 경우
- 두 개의 매개변수 a, b를 받며, 이때 [a,b]범위에서 균등한 확률을 가짐. 




- 확률밀도함수에서 면적이 확률이며, 면적은 1
- (b - a)*K = 1
- k =1 / (b-a)

- 범위 밖에서는 면적은 0


```python
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns

sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(10,6)})
```

```python
from scipy.stats import uniform
number = 10000
start = 20
width = 25

uniform_data = uniform.rvs(size=number, loc=start, scale=width) ## loc와 scale 사이의 균일한 연속변수 생성
axis = sns.distplot(uniform_data, bins=100, kde=True, color='skyblue', hist_kws={"linewidth": 15})
axis.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

# 4. Normal distribution (정규분포, 가우시안 분포)
- 정규 분포는 평균에 대해 대칭인 밀도 곡선을 가지고 있으며, 일반적으로 표준 편차로 정의

```python
from scipy.stats import norm

normal_data = norm.rvs(size=90000,loc=20,scale=30)
axis = sns.distplot(normal_data, bins=100, kde=True, color='skyblue', hist_kws={"linewidth": 15,'alpha':0.568})
axis.set(xlabel='Normal Distribution', ylabel='Frequency')
```

# 5. Exponential distribution (지수분포)
- 푸아송분포에서 유도된 분포
- 사건이 서로 독립적일 때, 일정 시간동안 발생하는 사건의 횟수가 푸아송 분포를 따른다면, 다음 사건이 일어날 때까지 대기 시간은 지수분포를 따른다


```python
# Exponential distribution
from scipy.stats import expon

expon_data = expon.rvs(scale=1,loc=0,size=1000)
axis = sns.distplot(expon_data, kde=True, bins=100, color='skyblue', hist_kws={"linewidth": 15})
axis.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

# 6. Binomial distribution(이항분포)
 - 성공.실패의 두 가지 가능한 결과만 존재하는 분포, 각 시행은 서로 독립적

```python
from scipy.stats import binom

binomial_data = binom.rvs(n=10, p=0.8,size=10000)

axis = sns.distplot(binomial_data, kde=False, color='red', hist_kws={"linewidth": 15})
axis.set(xlabel='Binomial Distribution', ylabel='Frequency')
```
